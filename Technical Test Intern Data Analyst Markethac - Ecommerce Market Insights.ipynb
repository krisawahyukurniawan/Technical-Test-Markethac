# Business Understanding

## Latar Belakang
Dalam beberapa tahun terakhir, industri ecommerce di Indonesia menunjukkan pertumbuhan yang sangat pesat, terutama pada kategori produk kecantikan seperti Sunscreen, Bedak, dan Make Up Remover. Dengan meningkatnya kesadaran masyarakat terhadap pentingnya perawatan kulit dan penampilan, permintaan terhadap produk-produk ini terus bertumbuh.
Namun, persaingan di pasar ini juga semakin ketat, dengan banyaknya brand baru yang bermunculan serta berbagai channel penjualan online yang terus berkembang. Untuk dapat memenangkan persaingan, brand perlu memahami perilaku pasar secara lebih mendalam, termasuk tren penjualan, pola diskon, preferensi channel, dan wilayah dengan potensi tinggi.

## Pernyataan Masalah
Brand yang ingin berinvestasi di kategori Sunscreen, Bedak, dan Make Up Remover perlu mengetahui:

1. Produk atau tipe listing apa yang memiliki performa penjualan terbaik.

2. Channel penjualan mana yang paling efektif untuk menjangkau konsumen.

3. Strategi harga dan diskon seperti apa yang paling efektif untuk meningkatkan penjualan.

4. Wilayah (kota) mana yang menunjukkan potensi pasar terbesar. Tanpa pemahaman mendalam ini, brand berisiko melakukan investasi yang tidak optimal, yang bisa menyebabkan rendahnya tingkat penjualan dan ROI.

## Tujuan
- Menganalisis data ecommerce listing untuk kategori Sunscreen, Bedak, dan Make Up Remover guna mengidentifikasi tren dan pola penting.

- Memberikan rekomendasi berbasis data terkait strategi harga, diskon, dan channel penjualan untuk brand yang ingin berinvestasi di kategori tersebut.

- Mengungkapkan peluang pasar potensial berdasarkan performa produk, channel, dan wilayah geografis.

- Membantu brand membuat keputusan investasi yang lebih efektif dan meningkatkan peluang keberhasilan di pasar ecommerce.

Tren peningkatan penjualan dengan diskon yang lebih tinggi menunjukkan elastisitas harga positif, namun setelah mencapai diskon sangat tinggi, dampak pada penjualan mulai stabil dan tidak sebanding dengan diskon yang diberikan. Oleh karena itu, diskon ekstrem (70-100%) harus dihindari karena tidak meningkatkan penjualan secara proporsional dan dapat merusak persepsi nilai produk.
4. Berdasarkan analisis, strategi bisnis yang tepat akan berfokus pada wilayah-wilayah dengan konsentrasi penjualan terbesar, khususnya Jakarta Timur yang memimpin penjualan. Mengingat dominasi wilayah Jabodetabek dalam top 10 kota dengan penjualan tertinggi, penting untuk mengalokasikan lebih banyak sumber daya dan promosi di daerah ini. Namun, perlu diperhatikan juga adanya kesenjangan signifikan dengan kota-kota besar lainnya seperti Surabaya, yang dapat menjadi peluang untuk memperkenalkan promosi atau kampanye yang dapat meningkatkan penjualan di luar Jakarta. Selain itu, disparitas antara frekuensi transaksi dan nilai penjualan di beberapa daerah menunjukkan adanya peluang untuk meningkatkan nilai rata-rata transaksi, terutama di kota-kota yang memiliki frekuensi transaksi tinggi namun nilai penjualan rendah, dengan menawarkan bundling produk, diskon spesial, atau peningkatan layanan pelanggan.

Strategi bisnis yang efektif harus fokus pada produk dengan rating tinggi (4-5 bintang), karena produk dengan rating ini mendominasi penjualan, serta mengoptimalkan jumlah review untuk meningkatkan social proof, terutama dengan mencapai lebih dari 50,000 review. Meskipun jumlah review berpengaruh, penambahan review setelah mencapai titik tertentu tidak memberikan dampak signifikan terhadap penjualan, sehingga kualitas dan pengalaman pelanggan lebih penting. Produk baru atau niche dengan sedikit review tetap memiliki potensi penjualan tinggi, yang menunjukkan perlunya pemasaran yang tepat. Selain itu, faktor-faktor lain seperti kualitas produk, harga, dan strategi pemasaran juga memainkan peran penting dalam mencapai penjualan tinggi, sehingga pendekatan yang holistik dan seimbang akan memberikan hasil terbaik.
